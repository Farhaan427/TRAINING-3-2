```python
import pandas as pd
data = {
    "Testcase": ["TC01", "TC02", "TC03", "TC04", "TC05"],
    "Status": ["PASS", "FAIL", "PASS", "FAIL", "PASS"]
}
df = pd.DataFrame(data)
df["Result"] = df["Status"].apply(lambda x: 1 if x == "PASS" else 0)
pass_percentage = (df["Result"].sum() / len(df)) * 100
print(df)
print("Pass Percentage:", pass_percentage)
```
