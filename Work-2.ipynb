```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "Name": ["Ashish", "Chirag", "Ankit", "Neha"],
    "Age": [80, np.nan, 70, np.nan],
    "Salary": [70000, 80000, np.nan, np.nan]
})
print(df)
```

```python
print(df.isnull().sum())
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
print(df)
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].median())
```

```python
print(df)
```
