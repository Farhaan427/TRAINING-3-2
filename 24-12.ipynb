## Question 1: Sentence and Word Tokenization

```python
from nltk.tokenize import sent_tokenize, word_tokenize

text = "Artificial Intelligence is transforming the world. NLP is a key area of AI."

sentences = sent_tokenize(text)
words = word_tokenize(text)

print("Sentence Tokens:", sentences)
print("Word Tokens:", words)
```

## Question 2: Stopword Removal

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words and w.isalpha()]

print(filtered_words)
```

## Question 3: Stemming

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
for w in filtered_words:
    print(w, '->', ps.stem(w))
```

## Question 4: Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
for w in filtered_words:
    print(w, '->', lemmatizer.lemmatize(w))
```

## Question 5: POS Tagging

```python
sentence = "Students are learning Natural Language Processing"
tokens = word_tokenize(sentence)
print(nltk.pos_tag(tokens))
```

## Question 6: Named Entity Recognition

```python
sentence = "Sundar Pichai is the CEO of Google and lives in California."
tokens = word_tokenize(sentence)
pos_tags = nltk.pos_tag(tokens)
print(nltk.ne_chunk(pos_tags))
```

## Question 7: Frequency Distribution

```python
from nltk.probability import FreqDist

text = "Data science uses data to extract meaningful insights from data"
tokens = word_tokenize(text.lower())
fd = FreqDist(tokens)
print(fd.most_common(5))
```
